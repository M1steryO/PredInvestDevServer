```python
def add_sma_ema(df, window=10):
    if 'Close' not in df.columns:
        print("В DataFrame нет столбца 'Close'.")
        return df
    df[f'SMA_{window}'] = df['Close'].rolling(window=window).mean()
    df[f'EMA_{window}'] = df['Close'].ewm(span=window, adjust=False).mean()
    return df

tickers = [
    'AFLT', 'ALRS', 'CHMF', 'GAZP', 'GMKN',
    'LKOH', 'MGNT', 'MOEX', 'MTSS', 'NVTK',
    'ROSN', 'SBER', 'TATN', 'VTBR', 'YNDX'
]
for ticker in tickers:
    var_name = f"df_{ticker}_daily"
    df_daily = globals()[var_name]
    df_daily = add_sma_ema(df_daily, window=10)
    df_daily = add_sma_ema(df_daily, window=50)
    globals()[var_name] = df_daily

for ticker in tickers:
    var_name = f"df_{ticker}_hourly"
    df_hourly = globals()[var_name]
    df_hourly = add_sma_ema(df_hourly, window=10)        
    df_hourly = add_sma_ema(df_hourly, window=50)  
    globals()[var_name] = df_hourly
      
```

```python
def calculate_atr(df, window=14):
    """
    Рассчитывает ATR (Average True Range) для DataFrame с колонками 'High', 'Low' и 'Close'.
    Предполагается, что данные отсортированы по 'Date'.
    """

    df['prev_close'] = df['Close'].shift(1)
    df['tr1'] = df['High'] - df['Low']
    df['tr2'] = (df['High'] - df['prev_close']).abs()
    df['tr3'] = (df['Low'] - df['prev_close']).abs()
    df['true_range'] = df[['tr1', 'tr2', 'tr3']].max(axis=1)
    
    # ATR – скользящее среднее истинного диапазона
    df[f'ATR_{window}'] = df['true_range'].rolling(window=window, min_periods=window).mean()
    
    # Удаляем промежуточные столбцы
    df.drop(['prev_close', 'tr1', 'tr2', 'tr3', 'true_range'], axis=1, inplace=True)
    return df


for ticker in tickers:
    var_name_hourly = f"df_{ticker}_hourly"
    if var_name_hourly in globals():
        df_hourly = globals()[var_name_hourly]
        df_hourly = calculate_atr(df_hourly, window=14)
        globals()[var_name_hourly] = df_hourly
        print(f"ATR_14 добавлен в {ticker} (часовой датасет).")
    else:
        print(f"Часовой датасет для {ticker} не найден.")
    var_name = f"df_{ticker}_daily" 
    if var_name in globals():
        df = globals()[var_name]
        df = calculate_atr(df, window=14)
        globals()[var_name] = df
        print(f"ATR_14 добавлен в {ticker} (дневной датасет).")
    else:
        print(f"Дневной датасет для {ticker} не найден.")
```

```python
import numpy as np


def add_volume_indicators(df):
    """
    Добавляет в DataFrame индикаторы OBV и VWAP.
    Предполагается, что в df присутствуют столбцы 'Close', 'High', 'Low' и 'Volume'.
    """
    # Считаем OBV:
    # Сдвигаем Close для сравнения с предыдущим значением
    df['prev_close'] = df['Close'].shift(1)
    # Если Close > prev_close, добавляем объем, если меньше — вычитаем, иначе 0.
    df['OBV'] = np.where(df['Close'] > df['prev_close'], df['Volume'],
                         np.where(df['Close'] < df['prev_close'], -df['Volume'], 0))
    # Кумулятивная сумма по OBV
    df['OBV'] = df['OBV'].cumsum()
    
    # Считаем VWAP:
    # Определяем типичную цену как среднее значение High, Low и Close
    typical_price = (df['High'] + df['Low'] + df['Close']) / 3
    # Кумулятивная сумма (типичная цена * объем)
    cumulative_tp_vol = (typical_price * df['Volume']).cumsum()
    # Кумулятивный объем
    cumulative_vol = df['Volume'].cumsum()
    # VWAP
    df['VWAP'] = cumulative_tp_vol / cumulative_vol
    
    # Удаляем временный столбец
    df.drop(columns=['prev_close'], inplace=True)
    
    return df



for ticker in tickers:

    daily_var = f"df_{ticker}_daily"
    if daily_var in globals():
        df_daily = globals()[daily_var]
        df_daily = add_volume_indicators(df_daily)
        globals()[daily_var] = df_daily
        print(f"[{ticker}] Дневной датасет обновлен: OBV и VWAP добавлены.")
    else:
        print(f"[{ticker}] Дневной датасет не найден.")
    
    hourly_var = f"df_{ticker}_hourly"
    if hourly_var in globals():
        df_hourly = globals()[hourly_var]
        df_hourly = add_volume_indicators(df_hourly)
        globals()[hourly_var] = df_hourly
        print(f"[{ticker}] Часовой датасет обновлен: OBV и VWAP добавлены.")
    else:
        print(f"[{ticker}] Часовой датасет не найден.")
```

```python
def add_rolling_std(df, window=20):
    """
    Добавляет в DataFrame новый столбец 'rolling_std_{window}', который
    представляет собой скользящее стандартное отклонение цены закрытия (Close)
    за заданное окно (window).
    """
    if 'Close' not in df.columns:
        print("Столбец 'Close' отсутствует в DataFrame")
        return df
    df[f'rolling_std_{window}'] = df['Close'].rolling(window=window, min_periods=window).std()
    return df




for ticker in tickers:
    var_name_daily = f"df_{ticker}_daily"
    if var_name_daily in globals():
        df_daily = globals()[var_name_daily]
        df_daily = add_rolling_std(df_daily, window=20)
        globals()[var_name_daily] = df_daily
        print(f"[{ticker}] В дневном датасете добавлен столбец rolling_std_20.")
    else:
        print(f"[{ticker}] Дневной датасет не найден.")


for ticker in tickers:
    var_name_hourly = f"df_{ticker}_hourly"
    if var_name_hourly in globals():
        df_hourly = globals()[var_name_hourly]
        df_hourly = add_rolling_std(df_hourly, window=20)
        globals()[var_name_hourly] = df_hourly
        print(f"[{ticker}] В часовом датасете добавлен столбец rolling_std_20.")
    else:
        print(f"[{ticker}] Часовой датасет не найден.")
```

```python
import requests
import xml.etree.ElementTree as ET
import datetime


# Функция для получения курса USD/RUB на заданную дату (используя ЦБ РФ)
def get_daily_usd_rate(date_req):
    """
    Получает курс доллара (USD/RUB) для заданной даты (объект date).
    """
    date_str = date_req.strftime("%d/%m/%Y")
    url = f"https://www.cbr.ru/scripts/XML_daily.asp?date_req={date_str}"
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()
    except Exception as e:
        print(f"Ошибка запроса для {date_str}: {e}")
        return None

    try:
        root = ET.fromstring(response.content)
    except Exception as e:
        print(f"Ошибка парсинга XML для {date_str}: {e}")
        return None

    for valute in root.findall('Valute'):
        if valute.find('CharCode').text == 'USD':
            value_text = valute.find('Value').text
            try:
                rate = float(value_text.replace(',', '.'))
                return rate
            except Exception as e:
                print(f"Ошибка преобразования курса {value_text} для {date_str}: {e}")
                return None
    return None



all_dates = set()
for ticker in tickers:
    var_name = f"df_{ticker}_daily"
    if var_name in globals():
        df = globals()[var_name]
        df['Date'] = pd.to_datetime(df['Date'], errors='coerce').dt.date
        all_dates.update(df['Date'].unique())
    else:
        print(f"Дневной датасет для {ticker} не найден.")

all_dates = sorted(all_dates)
print(f"Всего уникальных дат: {len(all_dates)}")

rates_list = []
for d in all_dates:
    usd_rate = get_daily_usd_rate(d)
    rates_list.append({'Date': d, 'USD_RUB': usd_rate})


df_rates = pd.DataFrame(rates_list)
df_rates['Date'] = pd.to_datetime(df_rates['Date'])
print("Пример данных по курсам USD/RUB:")
print(df_rates.head())

for ticker in tickers:
    var_name = f"df_{ticker}_daily"
    if var_name in globals():
        df = globals()[var_name]
        df['Date'] = pd.to_datetime(df['Date'])
        df_merged = pd.merge(df, df_rates, on="Date", how="left")
        globals()[var_name] = df_merged
        print(f"{ticker}: добавлен столбец USD_RUB. Всего строк: {len(df_merged)}")
    else:
        print(f"Дневной датасет для {ticker} не найден.")
```

```python
import pandas as pd
import numpy as np

def compute_RSI(df, window=14):
    """
    Рассчитывает RSI (Relative Strength Index) для DataFrame,
    в котором должен быть столбец 'Close'.
    
    Параметры:
      df: DataFrame с данными.
      window: число периодов для расчёта RSI (по умолчанию 14).
      
    Возвращает DataFrame с добавленным столбцом 'RSI'.
    """

    delta = df['Close'].diff()
    gain = delta.copy()
    loss = delta.copy()
    gain[gain < 0] = 0
    loss[loss > 0] = 0
    loss = loss.abs()

    # Рассчитываем средние значения прироста и потерь ( метод с rolling window)
    avg_gain = gain.rolling(window=window, min_periods=window).mean()
    avg_loss = loss.rolling(window=window, min_periods=window).mean()

    # Вычисляем относительную силу (RS)
    rs = avg_gain / avg_loss
    # Вычисляем RSI
    rsi = 100 - (100 / (1 + rs))
    df['RSI'] = rsi

    return df


# Применяем вычисление RSI для каждого тикера
for ticker in tickers:
    var_name = f"df_{ticker}_daily"
    if var_name in globals():
        df = globals()[var_name]
    
        df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
        # Рассчитываем RSI и добавляем столбец 'RSI'
        df = compute_RSI(df, window=14)
        globals()[var_name] = df
        print(f"{ticker}: RSI добавлен. Всего строк: {len(df)}")
    else:
        print(f"Дневной датасет для {ticker} не найден.")
var_name = f"df_YNDX_daily"
if var_name in globals():
    df = globals()[var_name]
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
        # Рассчитываем RSI и добавляем столбец 'RSI'        
    df = compute_RSI(df, window=14)
    globals()[var_name] = df
        
```

```python

from ta.trend import ADXIndicator
from ta.momentum import StochasticOscillator

def add_additional_features(df):
    # 1. Daily Return: процентное изменение цены закрытия
    df['DailyReturn'] = df['Close'].pct_change()
    
    # 2. ADX_14: сила тренда
    adx_ind = ADXIndicator(high=df['High'], low=df['Low'], close=df['Close'], window=14, fillna=True)
    df['ADX_14'] = adx_ind.adx()
    
    # 3. Stochastic Oscillator (%K и %D)
    stoch = StochasticOscillator(high=df['High'], low=df['Low'], close=df['Close'], window=14, smooth_window=3, fillna=True)
    df['Stoch_%K'] = stoch.stoch()
    df['Stoch_%D'] = stoch.stoch_signal()
    
    return df


for ticker in tickers:
    var_name = f"df_{ticker}_daily"
    if var_name in globals():
        df = globals()[var_name]
        # Убедимся, что 'Date' приведен к datetime (это не критично для расчёта, но полезно для последующего объединения)
        df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
        # Добавляем новые признаки
        df = add_additional_features(df)
        globals()[var_name] = df
        print(f"{ticker}: добавлены DailyReturn, ADX_14, Stoch_%K и Stoch_%D. Всего строк: {len(df)}")
    else:
        print(f"Дневной датасет для {ticker} не найден.")
```

```python

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler


aggregated_features = {}

for ticker in tickers:
    var_name = f"df_{ticker}_daily"
    if var_name in globals():
        df = globals()[var_name].copy()
        df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
        df.sort_values('Date', inplace=True)
        
        
        df['Return'] = df['Close'].pct_change()
        mean_return    = df['Return'].mean()                # Средняя дневная доходность
        volatility     = df['Return'].std()                 # Волатильность (std дневных доходностей)
        mean_atr       = df['ATR_14'].mean() if 'ATR_14' in df.columns else np.nan
        mean_macd_diff = df['MACD_diff'].mean() if 'MACD_diff' in df.columns else np.nan
        mean_volume    = df['Volume'].mean() if 'Volume' in df.columns else np.nan
        
        aggregated_features[ticker] = [mean_return, volatility, mean_atr, mean_macd_diff, mean_volume]
    else:
        print(f"Дневной датасет для {ticker} не найден.")


df_features = pd.DataFrame.from_dict(aggregated_features, orient='index',
                                     columns=['MeanReturn', 'Volatility', 'MeanATR', 'MeanMACD_diff', 'MeanVolume'])
df_features.reset_index(inplace=True)
df_features.rename(columns={'index': 'Ticker'}, inplace=True)

print("Агрегированные признаки по тикерам:")
print(df_features)

# Нормализуем признаки, чтобы они были на одном масштабе
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_features[['MeanReturn', 'Volatility', 'MeanATR', 'MeanMACD_diff', 'MeanVolume']])

# Метод локтя: рассчитываем инерцию (within-cluster sum of squares) для k от 1 до 10
inertias = []
K_range = range(1, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)


plt.figure(figsize=(8, 5))
plt.plot(K_range, inertias, marker='o', linestyle='-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia (Within-cluster Sum of Squares)')
plt.title('Elbow Method for Optimal k')
plt.xticks(K_range)
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np

# Функция для расчёта Williams %R за заданное окно (например, 14 дней)
def compute_williams_r(df, window=14):
    """
    Вычисляет Williams %R по формуле:
      %R = -100 * (Highest High - Close) / (Highest High - Lowest Low)
    за заданное окно.
    """
    highest_high = df['High'].rolling(window=window, min_periods=window).max()
    lowest_low = df['Low'].rolling(window=window, min_periods=window).min()
    # Если диапазон равен 0, вернем NaN
    williams_r = -100 * (highest_high - df['Close']) / (highest_high - lowest_low)
    return williams_r

# Функция для расчёта Relative Volume
def compute_relative_volume(df, window=20):
    volume_sma = df['Volume'].rolling(window=window, min_periods=window).mean()
    relative_volume = df['Volume'] / volume_sma
    return relative_volume

# Извлекаем тикеры из кластера 2 
cluster2_tickers = df_features.loc[df_features['Cluster'] == 2, 'Ticker'].tolist()
print("Тикеры из кластера 2:", cluster2_tickers)

# Для каждого тикера из кластера 2 добавляем новые признаки
for ticker in cluster2_tickers:
    var_name = f"df_{ticker}_daily"
    if var_name in globals():
        df = globals()[var_name].copy()
        # Убедимся, что данные отсортированы по дате
        df = df.sort_values('Date').reset_index(drop=True)
        # Вычисляем Williams %R (окно 14 дней)
        df['Williams_%R'] = compute_williams_r(df, window=14)
        # Вычисляем Relative Volume (относительный объем за 20 дней)
        df['RelativeVolume'] = compute_relative_volume(df, window=20)
        globals()[var_name] = df
        print(f"{ticker}: добавлены признаки 'Williams_%R' и 'RelativeVolume'.")
    else:
        print(f"Дневной датасет для {ticker} не найден.")
```

```python
# Функция Williams %R для часового датасета
def compute_williams_r_hourly(df, window=14):
    highest_high = df['High'].rolling(window=window, min_periods=window).max()
    lowest_low = df['Low'].rolling(window=window, min_periods=window).min()
    return -100 * (highest_high - df['Close']) / (highest_high - lowest_low)

# Функция наклона RSI (RSI_slope) по часовым данным
def compute_rsi_slope_hourly(series, window=5):
    def slope(x):
        if np.isnan(x).any():
            return np.nan
        return np.polyfit(np.arange(len(x)), x, 1)[0]
    return series.rolling(window=window, min_periods=window).apply(slope, raw=True)

for ticker in tickers:
    var_name = f"df_{ticker}_hourly"
    if var_name in globals():
        df = globals()[var_name].copy()
        df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

        if 'ticker' not in df.columns:
            df['ticker'] = ticker
        # 1. RSI_lag1_hourly: сдвиг RSI на 1 период (1 час)
        if 'RSI' in df.columns:
            df['RSI_lag1_hourly'] = df['RSI'].shift(1)
        else:
            print(f"{ticker}: столбец 'RSI' не найден в часовом датасете.")
        
        # 2. Stoch_%D_hourly
        if 'Stoch_%D' not in df.columns:
           
            if 'Stoch_%K' in df.columns:
                df['Stoch_%D_hourly'] = df['Stoch_%K'].rolling(window=3, min_periods=3).mean()
            else:
                print(f"{ticker}: для расчета Stoch_%D_hourly отсутствует 'Stoch_%K'.")
        else:
           
            df.rename(columns={'Stoch_%D': 'Stoch_%D_hourly'}, inplace=True)
        
        # 3. Williams_%R_hourly
        df['Williams_%R_hourly'] = compute_williams_r_hourly(df, window=14)
        
        # 4. Price_deviation_SMA50_hourly
        if 'SMA_50' not in df.columns:
            df['SMA_50'] = df['Close'].rolling(window=50, min_periods=50).mean()
        df['Price_deviation_SMA50_hourly'] = (df['Close'] - df['SMA_50']) / df['SMA_50']
        
        # 5. MACD_hourly
   
        if 'MACD' in df.columns:
            df.rename(columns={'MACD': 'MACD_hourly'}, inplace=True)
        
        # 6. RSI_slope_hourly
        if 'RSI' in df.columns:
            df['RSI_slope_hourly'] = compute_rsi_slope_hourly(df['RSI'], window=5)
        
        main_features_hourly = ['RSI_lag1_hourly', 'Stoch_%D_hourly', 'Williams_%R_hourly',
                                'Price_deviation_SMA50_hourly', 'MACD_hourly', 'RSI_slope_hourly']
        
        available_features = [feat for feat in main_features_hourly if feat in df.columns]
        df_hourly_main = df[['Date', 'ticker'] + available_features].copy()
        
        globals()[f"df_{ticker}_hourly_main"] = df_hourly_main
        print(f"{ticker}: создан датасет с основными часовыми признаками (shape: {df_hourly_main.shape}).")
    else:
        print(f"{ticker}: часовой датасет не найден.")
```

```python
import os



# Функция для вычисления Williams %R для часового датасета
def compute_williams_r_hourly(df, window=14):
    highest_high = df['High'].rolling(window=window, min_periods=window).max()
    lowest_low = df['Low'].rolling(window=window, min_periods=window).min()
    return -100 * (highest_high - df['Close']) / (highest_high - lowest_low)

# Функция для вычисления наклона RSI (RSI_slope_hourly) 
def compute_rsi_slope_hourly(series, window=5):
    def slope(x):
        if np.isnan(x).any():
            return np.nan
        return np.polyfit(np.arange(len(x)), x, 1)[0]
    return series.rolling(window=window, min_periods=window).apply(slope, raw=True)


cluster2_tickers = df_features.loc[df_features['Cluster'] == 2, 'Ticker'].tolist()
print("Tickers in Cluster 2:", cluster2_tickers)


base_columns = ['Date', 'ticker', 'Open', 'Close', 'High', 'Low', 'Volume', 'value', 'end', 
                'SMA_10', 'EMA_10', 'SMA_50', 'EMA_50', 'MACD', 'MACD_signal', 'MACD_diff', 
                'ATR_14', 'OBV', 'VWAP', 'rolling_std_20', 'USD_RUB', 'Brent', 'Inflation', 'RSI']
new_hourly_features = ['RSI_lag1_hourly', 'RSI_lag2_hourly', 'Stoch_%D_hourly', 
                       'Williams_%R_hourly', 'Price_deviation_SMA50_hourly', 'RSI_slope_hourly',
                       'HourlyReturn', 'Momentum_5_hourly', 'RollingVol_10_hourly', 'RelativeVolume_hourly']
all_cols = base_columns + new_hourly_features

os.makedirs("datasets_claster2_hourly", exist_ok=True)

for ticker in cluster2_tickers:
    var_name = f"df_{ticker}_hourly"
    if var_name in globals():
        df = globals()[var_name].copy()
        df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
        df = df.sort_values('Date').reset_index(drop=True)
        if 'ticker' not in df.columns:
            df['ticker'] = ticker
        
        # 1. RSI_lag1_hourly и RSI_lag2_hourly
        if 'RSI' in df.columns:
            df['RSI_lag1_hourly'] = df['RSI'].shift(1)
            df['RSI_lag2_hourly'] = df['RSI'].shift(2)
        else:
            print(f"{ticker}: 'RSI' not found for lag computation.")
        
        # 2. Stoch_%D_hourly: если отсутствует, вычисляем как 3-hour SMA от 'Stoch_%K'
        if 'Stoch_%D_hourly' not in df.columns:
            if 'Stoch_%K' in df.columns:
                df['Stoch_%D_hourly'] = df['Stoch_%K'].rolling(window=3, min_periods=3).mean()
            else:
                print(f"{ticker}: 'Stoch_%K' not found for computing Stoch_%D_hourly.")
        
        # 3. Williams_%R_hourly
        if all(col in df.columns for col in ['High', 'Low', 'Close']):
            df['Williams_%R_hourly'] = compute_williams_r_hourly(df, window=14)
        else:
            print(f"{ticker}: Missing 'High', 'Low', or 'Close' for Williams_%R_hourly.")
        
        # 4. Price_deviation_SMA50_hourly: рассчитываем 50-часовую SMA, если отсутствует
        if 'SMA_50' not in df.columns:
            df['SMA_50'] = df['Close'].rolling(window=50, min_periods=50).mean()
        df['Price_deviation_SMA50_hourly'] = (df['Close'] - df['SMA_50']) / df['SMA_50']
        
        # 5. MACD_hourly: если 'MACD' существует, переименовываем его для ясности
        if 'MACD' in df.columns and 'MACD_hourly' not in df.columns:
            df.rename(columns={'MACD': 'MACD_hourly'}, inplace=True)
        
        # 6. RSI_slope_hourly: вычисляем наклон RSI по окну 5 часов
        if 'RSI' in df.columns:
            df['RSI_slope_hourly'] = compute_rsi_slope_hourly(df['RSI'], window=5)
        
        # 7. HourlyReturn: процентное изменение цены закрытия между часами
        df['HourlyReturn'] = df['Close'].pct_change()
        
        # 8. Momentum_5_hourly: изменение цены закрытия по сравнению с 5 часов назад
        df['Momentum_5_hourly'] = (df['Close'] - df['Close'].shift(5)) / df['Close'].shift(5)
        
        # 9. RollingVol_10_hourly: стандартное отклонение HourlyReturn за 10 часов
        df['RollingVol_10_hourly'] = df['HourlyReturn'].rolling(window=10, min_periods=10).std()
        
        # 10. RelativeVolume_hourly: текущий объем / 20-часовая SMA объёма
        df['Volume_SMA20_hourly'] = df['Volume'].rolling(window=20, min_periods=20).mean()
        df['RelativeVolume_hourly'] = df['Volume'] / df['Volume_SMA20_hourly']
        # df.drop(columns=['Volume_SMA20_hourly'], inplace=True)
        
        final_cols = [col for col in all_cols if col in df.columns]
        df_hourly_main = df[final_cols].copy()
        
        filename = os.path.join("datasets_claster2_hourly", f"{ticker}_hourly.csv")
        df_hourly_main.to_csv(filename, index=False)
        print(f"{ticker}: Hourly dataset saved to {filename} (shape: {df_hourly_main.shape}).")
    else:
        print(f"{ticker}: Hourly dataset {var_name} not found.")
```

```python
import os

def compute_RSI(series, window=14):
    """
    Вычисляет RSI (Relative Strength Index) для заданной серии (например, цены закрытия).
    window - число периодов (по умолчанию 14).
    """
    delta = series.diff()
    gain = delta.clip(lower=0)
    loss = -delta.clip(upper=0)
    
    # Скользящие средние прироста и потерь
    avg_gain = gain.rolling(window=window, min_periods=window).mean()
    avg_loss = loss.rolling(window=window, min_periods=window).mean()
    
    # Относительная сила (RS)
    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    return rsi


hourly_folder = "datasets_claster2_hourly"
cluster2_tickers = ["AFLT", "ALRS", "GAZP", "GMKN", "MOEX", "MTSS", "SBER", "VTBR"]

for ticker in cluster2_tickers:
    csv_path = os.path.join(hourly_folder, f"{ticker}_hourly.csv")
    if os.path.exists(csv_path):
      
        df = pd.read_csv(csv_path)
        
       
        if 'Date' in df.columns:
            df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
            df.sort_values('Date', inplace=True)
            df.reset_index(drop=True, inplace=True)
        

        if 'RSI' not in df.columns:
            df['RSI'] = compute_RSI(df['Close'], window=14)
            print(f"{ticker}: RSI добавлен в часовой датасет.")
        else:
            print(f"{ticker}: RSI уже существует, пересчитываем...")
            df['RSI'] = compute_RSI(df['Close'], window=14)

        df.to_csv(csv_path, index=False)
        print(f"{ticker}: часовой датасет с RSI сохранен в {csv_path} (строк: {len(df)})")
    else:
        print(f"Файл {csv_path} не найден, пропускаем {ticker}.")
```
