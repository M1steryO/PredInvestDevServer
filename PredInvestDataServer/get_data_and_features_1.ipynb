```python
import pandas as pd
from moexalgo import Ticker


tickers = [
    'AFLT', 'ALRS', 'CHMF', 'GAZP', 'GMKN',
    'LKOH', 'MGNT', 'MOEX', 'MTSS', 'NVTK',
    'ROSN', 'SBER', 'TATN', 'VTBR'
]


start_date = "2015-01-05"
end_date   = "2024-12-30"

for ticker in tickers:
    print(f"Обрабатываем {ticker}...")
    try:
        ticker_obj = Ticker(ticker)
        
        # Получаем дневные свечи 
        df_daily = ticker_obj.candles(start=start_date, end=end_date, period='1d')

        if "begin" in df_daily.columns:
            df_daily["begin"] = pd.to_datetime(df_daily["begin"])
            df_daily = df_daily.rename(columns={"begin": "date"})
        
        globals()[f"df_{ticker}_daily"] = df_daily
        print(f"  Дневные данные для {ticker} сохранены в переменную df_{ticker}_daily")
    except Exception as e:
        print(f"  Ошибка при получении дневных данных для {ticker}: {e}")
    
    try:
        # Получаем почасовые свечи с помощью параметра period='1h'
        df_hourly = ticker_obj.candles(start=start_date, end=end_date, period='1h')
        if "begin" in df_hourly.columns:
            df_hourly["begin"] = pd.to_datetime(df_hourly["begin"])
            df_hourly = df_hourly.rename(columns={"begin": "date"})
        
        globals()[f"df_{ticker}_hourly"] = df_hourly
        print(f"  Почасовые данные для {ticker} сохранены в переменную df_{ticker}_hourly")
    except Exception as e:
        print(f"  Ошибка при получении почасовых данных для {ticker}: {e}")
```

```python
def compute_SMA_10(df):
    """
    Вычисляет простое скользящее среднее (SMA) с периодом 10 для столбца 'Close'
    и добавляет его в DataFrame как новый столбец 'SMA_10'.
    """
    df['SMA_10'] = df['Close'].rolling(window=10).mean()
    return df

def compute_EMA_10(df):
    """
    Вычисляет экспоненциальное скользящее среднее (EMA) с периодом 10 для столбца 'Close'.
    """
    df['EMA_10'] = df['Close'].ewm(span=10, adjust=False).mean()
    return df

def compute_SMA_50(df):
    """
    Вычисляет простое скользящее среднее (SMA) с периодом 50 для столбца 'Close'.
    """
    df['SMA_50'] = df['Close'].rolling(window=50).mean()
    return df

def compute_EMA_50(df):
    """
    Вычисляет экспоненциальное скользящее среднее (EMA) с периодом 50 для столбца 'Close'.
    """
    df['EMA_50'] = df['Close'].ewm(span=50, adjust=False).mean()
    return df

def apply_indicators(df):
    """
    Применяет все четыре индикатора (SMA_10, EMA_10, SMA_50, EMA_50) к DataFrame.
    """
    df = compute_SMA_10(df)
    df = compute_EMA_10(df)
    df = compute_SMA_50(df)
    df = compute_EMA_50(df)
    return df
```

```python
for ticker in tickers:
    var_name = f"df_{ticker}_daily"
    if var_name in globals():
        df = globals()[var_name]
        df_updated = apply_indicators(df.copy())
        globals()[var_name] = df_updated
        print(f"=== {ticker} дневные данные после применения индикаторов ===")
        display(df_updated.head(5))
    else:
        print(f"Дневные данные для тикера {ticker} не найдены!")
        
```

```python
def compute_MACD(df, short_span=12, long_span=26):
    """
    Вычисляет MACD как разницу между EMA(12) и EMA(26) для столбца 'Close'
    """
    ema_short = df['Close'].ewm(span=short_span, adjust=False).mean()
    ema_long = df['Close'].ewm(span=long_span, adjust=False).mean()
    return ema_short - ema_long

def compute_MACD_signal(macd_series, signal_span=9):
    """
    Вычисляет сигнальную линию MACD как EMA от MACD с периодом signal_span (по умолчанию 9).
    """
    return macd_series.ewm(span=signal_span, adjust=False).mean()

def compute_MACD_diff(macd_series, macd_signal_series):
    """
    Вычисляет разницу между MACD и его сигнальной линией (MACD_diff).
    """
    return macd_series - macd_signal_series

def apply_MACD_indicators(df, short_span=12, long_span=26, signal_span=9):
    """
    Применяет расчёт MACD, MACD_signal и MACD_diff к DataFrame (используется столбец 'Close'),
    добавляя новые столбцы: 'MACD', 'MACD_signal', 'MACD_diff'.
    """
    macd = compute_MACD(df, short_span, long_span)
    macd_signal = compute_MACD_signal(macd, signal_span)
    macd_diff = compute_MACD_diff(macd, macd_signal)
    
    df['MACD'] = macd
    df['MACD_signal'] = macd_signal
    df['MACD_diff'] = macd_diff
    return df
```

```python
for ticker in tickers:
    var_name = f"df_{ticker}_daily"
    if var_name in globals():
        df = globals()[var_name]
        df = apply_MACD_indicators(df)
        globals()[var_name] = df
        print(f"Первые 5 строк дневных данных для {ticker} после добавления индикаторов MACD:")
        display(df.head(5))
    else:
        print(f"Дневные данные для тикера {ticker} не найдены!")
```

```python
def compute_ATR_14(df):
    """
    Вычисляет ATR (Average True Range) за 14 периодов.
    True Range (TR) для каждого периода = max( High - Low,
                                             abs(High - предыдущий Close),
                                             abs(Low - предыдущий Close) ).
    """
    df = df.copy()
    df['prev_Close'] = df['Close'].shift(1)
    df['TR'] = df.apply(lambda row: max(
                            row['High'] - row['Low'],
                            abs(row['High'] - row['prev_Close']) if pd.notna(row['prev_Close']) else 0,
                            abs(row['Low'] - row['prev_Close']) if pd.notna(row['prev_Close']) else 0
                        ), axis=1)
    df['ATR_14'] = df['TR'].rolling(window=14).mean()
    df.drop(columns=['prev_Close', 'TR'], inplace=True)
    return df

def compute_OBV(df):
    """
    Вычисляет On-Balance Volume (OBV).
    """
    df = df.copy()
    obv = [0]  
    for i in range(1, len(df)):
        if df['Close'].iloc[i] > df['Close'].iloc[i-1]:
            obv.append(obv[-1] + df['Volume'].iloc[i])
        elif df['Close'].iloc[i] < df['Close'].iloc[i-1]:
            obv.append(obv[-1] - df['Volume'].iloc[i])
        else:
            obv.append(obv[-1])
    df['OBV'] = obv
    return df

def compute_VWAP(df, window=20):
    """
    Вычисляет VWAP (Volume Weighted Average Price) по скользящему окну.
    Для каждой строки рассчитываем типичную цену: (High+Low+Close)/3.
    Затем VWAP за окно = сумма(typical_price * Volume) / сумма(Volume).
    """
    df = df.copy()
    # Вычисляем типичную цену
    df['typical_price'] = (df['High'] + df['Low'] + df['Close']) / 3.0
    df['VWAP'] = (df['typical_price'] * df['Volume']).rolling(window=window).sum() / df['Volume'].rolling(window=window).sum()
    df.drop(columns=['typical_price'], inplace=True)
    return df

def compute_rolling_std_20(df):
    """
    Вычисляет скользящее стандартное отклонение от 'Close' за 20 периодов.
    """
    df = df.copy()
    df['rolling_std_20'] = df['Close'].rolling(window=20).std()
    return df

def apply_extra_indicators(df):
    """
    Применяет вычисление четырёх индикаторов:
      - ATR_14
      - OBV
      - VWAP (rolling VWAP с окном 20)
      - rolling_std_20
    """
    df = compute_ATR_14(df)
    df = compute_OBV(df)
    df = compute_VWAP(df, window=20)
    df = compute_rolling_std_20(df)
    return df


for ticker in tickers:
    var_name = f"df_{ticker}_daily"
    if var_name in globals():
        df = globals()[var_name]
        df = apply_extra_indicators(df)
        globals()[var_name] = df 
        print(f"Первые 5 строк дневных данных для {ticker} после добавления индикаторов ATR_14, OBV, VWAP и rolling_std_20:")
        display(df.head(5))
    else:
        print(f"Дневные данные для тикера {ticker} не найдены!")
```

```python
import yfinance as yf


def get_usd_rub_brent_data(start_date, end_date):
    """
    Загружает с Yahoo Finance данные:
      - USD/RUB (тикер "RUB=X") – курс обмена в рублях за 1 USD,
      - Brent (тикер "BZ=F") – цена нефти Brent.
    Затем формирует DataFrame с колонками:
      'Date', 'USD_RUB', 'Brent', 'Inflation'
    В колонке 'Inflation' значения заполняются NaN.
    """
    # Загрузка данных для USD/RUB; используем только 'Close'
    usd_rub = yf.download("RUB=X", start=start_date, end=end_date)
    usd_rub = usd_rub[['Close']].rename(columns={'Close': 'USD_RUB'})
    
    # Загрузка данных для Brent
    brent = yf.download("BZ=F", start=start_date, end=end_date)
    brent = brent[['Close']].rename(columns={'Close': 'Brent'})
    # Объединяем по индексу (дате)
    merged = pd.merge(usd_rub, brent, left_index=True, right_index=True, how='outer')
    # Добавляем колонку инфляции, заполненную NaN (можно использовать pd.NA или np.nan)
    merged['Inflation'] = pd.NA
    # Сбрасываем индекс, чтобы дата стала столбцом, и переименовываем его в 'Date'
    merged.reset_index(inplace=True)
    merged.rename(columns={'Date': 'Date'}, inplace=True)
    
    return merged

date_range_start = "2015-01-05"
date_range_end   = "2024-12-30"

usd_brent_data = get_usd_rub_brent_data(date_range_start, date_range_end)
usd_brent_data['Date'] = pd.to_datetime(usd_brent_data['Date'])

```

```python
def compute_RSI(df, period=14):
    """
    Вычисляет Relative Strength Index (RSI) за указанный период.
    """
    df = df.copy()
    delta = df['Close'].diff()
    gain = delta.clip(lower=0)
    loss = -delta.clip(upper=0)
    avg_gain = gain.rolling(window=period).mean()
    avg_loss = loss.rolling(window=period).mean()
    rs = avg_gain / avg_loss
    df['RSI'] = 100 - (100 / (1 + rs))
    return df

def compute_DailyReturn(df):
    """
    Вычисляет дневную доходность как процентное изменение цены закрытия.
    """
    df = df.copy()
    df['DailyReturn'] = df['Close'].pct_change() * 100
    return df

def compute_ADX_14(df):
    """
    Вычисляет ADX (Average Directional Index) за 14 периодов.
    Требуются столбцы: 'High', 'Low', 'Close'
    """
    df = df.copy()
    df['prev_Close'] = df['Close'].shift(1)
    df['prev_High'] = df['High'].shift(1)
    df['prev_Low'] = df['Low'].shift(1)
    
    df['TR'] = df[['High', 'Low', 'prev_Close']].apply(
        lambda row: max(row['High'] - row['Low'],
                        abs(row['High'] - row['prev_Close']),
                        abs(row['Low'] - row['prev_Close'])), axis=1)
    
    df['+DM'] = df['High'] - df['prev_High']
    df['-DM'] = df['prev_Low'] - df['Low']
    df['+DM'] = df.apply(lambda row: row['+DM'] if (row['+DM'] > row['-DM'] and row['+DM'] > 0) else 0, axis=1)
    df['-DM'] = df.apply(lambda row: row['-DM'] if (row['-DM'] > row['+DM'] and row['-DM'] > 0) else 0, axis=1)
    
    period = 14
    df['TR14'] = df['TR'].rolling(window=period).sum()
    df['+DM14'] = df['+DM'].rolling(window=period).sum()
    df['-DM14'] = df['-DM'].rolling(window=period).sum()
    df['DI+'] = 100 * (df['+DM14'] / df['TR14'])
    df['DI-'] = 100 * (df['-DM14'] / df['TR14'])
    df['DX'] = 100 * (abs(df['DI+'] - df['DI-']) / (df['DI+'] + df['DI-']))
    df['ADX_14'] = df['DX'].rolling(window=period).mean()
    
    # Удаляем временные столбцы
    df.drop(columns=['prev_Close', 'prev_High', 'prev_Low', 'TR', '+DM', '-DM', 'TR14', '+DM14', '-DM14', 'DI+', 'DI-', 'DX'], inplace=True)
    return df

def compute_Stoch_K(df, period=14):
    """
    Вычисляет %K стахастика за указанный период:
        %K = 100 * (Close - Min(Low)) / (Max(High) - Min(Low))
    """
    df = df.copy()
    df['Stoch_%K'] = 100 * (df['Close'] - df['Low'].rolling(window=period).min()) / (df['High'].rolling(window=period).max() - df['Low'].rolling(window=period).min())
    return df

def compute_Stoch_D(df, period=3):
    """
    Вычисляет %D стахастика как скользящее среднее %K за указанный период (по умолчанию 3).
    """
    df = df.copy()
    df['Stoch_%D'] = df['Stoch_%K'].rolling(window=period).mean()
    return df

def compute_RSI_lags(df):
    """
    Создаёт столбцы 'RSI_lag1' и 'RSI_lag2' со значениями RSI сдвинутыми на 1 и 2 периода соответственно.
    """
    df = df.copy()
    df['RSI_lag1'] = df['RSI'].shift(1)
    df['RSI_lag2'] = df['RSI'].shift(2)
    return df

def apply_all_indicators(df):
    """
    Применяет все индикаторы: RSI, DailyReturn, ADX_14, Stoch_%K, Stoch_%D, а затем добавляет сдвиги RSI_lag1 и RSI_lag2.
    """
    df = compute_RSI(df, period=14)
    df = compute_DailyReturn(df)
    df = compute_ADX_14(df)
    df = compute_Stoch_K(df, period=14)
    df = compute_Stoch_D(df, period=3)
    df = compute_RSI_lags(df)
    return df


for ticker in tickers:
    var_name = f"df_{ticker}_daily"
    if var_name in globals():
        df = globals()[var_name]
        df_updated = apply_all_indicators(df)
        globals()[var_name] = df_updated  
        print(f"Первые 5 строк дневных данных для {ticker} после применения всех индикаторов:")
        display(df_updated.head(5))
    else:
        print(f"Дневные данные для тикера {ticker} не найдены!")
```

```python
def compute_RSI(df, period=14):
    df = df.copy()
    delta = df['Close'].diff()
    gain = delta.clip(lower=0)
    loss = -delta.clip(upper=0)
    avg_gain = gain.rolling(window=period).mean()
    avg_loss = loss.rolling(window=period).mean()
    rs = avg_gain / avg_loss
    df['RSI'] = 100 - (100 / (1 + rs))
    return df

def compute_price_deviation_SMA50(df, window=50):
    """Вычисляет 50-дневную SMA и процентное отклонение Close от SMA50."""
    df = df.copy()
    df['SMA_50'] = df['Close'].rolling(window=window).mean()
    df['Price_deviation_SMA50'] = (df['Close'] - df['SMA_50']) / df['SMA_50'] * 100
    return df

def compute_SMA_20(df, window=20):
    """Вычисляет 20-дневную SMA по Close."""
    df = df.copy()
    df['SMA_20'] = df['Close'].rolling(window=window).mean()
    return df

def compute_RollingStd_20(df, window=20):
    """Вычисляет 20-дневное скользящее стандартное отклонение по Close."""
    df = df.copy()
    df['RollingStd_20'] = df['Close'].rolling(window=window).std()
    return df

def compute_Bollinger_BandWidth(df, window=20, num_std=2):
    """
    Рассчитывает Bollinger Band Width как:
        ( (SMA_20 + num_std*RollingStd_20) - (SMA_20 - num_std*RollingStd_20) ) / SMA_20 * 100
    """
    df = df.copy()
    sma20 = df['Close'].rolling(window=window).mean()
    rolling_std = df['Close'].rolling(window=window).std()
    upper_band = sma20 + num_std * rolling_std
    lower_band = sma20 - num_std * rolling_std
    df['Bollinger_BandWidth'] = (upper_band - lower_band) / sma20 * 100
    return df

def compute_VWAP_deviation(df, window=20):
    """
    Вычисляет VWAP (Volume Weighted Average Price) за заданное окно и процентное отклонение Close от VWAP.
    VWAP = sum(typical_price*Volume)/sum(Volume), где typical_price = (High+Low+Close)/3.
    """
    df = df.copy()
    typical_price = (df['High'] + df['Low'] + df['Close']) / 3.0
    rolling_vwap = (typical_price * df['Volume']).rolling(window=window).sum() / df['Volume'].rolling(window=window).sum()
    df['VWAP_deviation'] = (df['Close'] - rolling_vwap) / rolling_vwap * 100
    return df

def compute_RSI_slope(df, window=5):
    """
    Вычисляет наклон RSI по скользящему окну.
    Для каждого окна из значений RSI, проводит линейную регрессию (np.polyfit) по индексам и возвращает коэффициент наклона.
    """
    df = df.copy()
    def slope_func(rsi_series):
        if rsi_series.isnull().any():
            return np.nan
        x = np.arange(len(rsi_series))
        slope, _ = np.polyfit(x, rsi_series, 1)
        return slope
    df['RSI_slope'] = df['RSI'].rolling(window=window).apply(slope_func, raw=False)
    return df

def compute_Williams_R(df, period=14):
    """
    Вычисляет Williams %R за указанный период:  %R = -100 * (Highest High - Close)/(Highest High - Lowest Low)
    """
    df = df.copy()
    highest_high = df['High'].rolling(window=period).max()
    lowest_low = df['Low'].rolling(window=period).min()
    df['Williams_%R'] = -100 * (highest_high - df['Close']) / (highest_high - lowest_low)
    return df

def compute_RelativeVolume(df, window=20):
    """
    Вычисляет Relative Volume как текущий Volume делённый на 20-дневное среднее Volume.
    """
    df = df.copy()
    avg_volume = df['Volume'].rolling(window=window).mean()
    df['RelativeVolume'] = df['Volume'] / avg_volume
    return df

def apply_new_indicators(df):
    """Применяет все новые индикаторы к DataFrame и возвращает обновлённый результат."""
    df = df.copy()
    if 'RSI' not in df.columns:
        df = compute_RSI(df, period=14)
    df = compute_price_deviation_SMA50(df, window=50)
    df = compute_SMA_20(df, window=20)
    df = compute_RollingStd_20(df, window=20)
    df = compute_Bollinger_BandWidth(df, window=20, num_std=2)
    df = compute_VWAP_deviation(df, window=20)
    df = compute_RSI_slope(df, window=5)
    df = compute_Williams_R(df, period=14)
    df = compute_RelativeVolume(df, window=20)
    return df



for ticker in tickers:
    var_name = f"df_{ticker}_daily"
    if var_name in globals():
        df = globals()[var_name]
        df_updated = apply_new_indicators(df)
        globals()[var_name] = df_updated 
        print(f"Первые 5 строк дневных данных для {ticker} после добавления новых индикаторов:")
        display(df_updated.head(5))
    else:
        print(f"Дневные данные для тикера {ticker} не найдены!")
```

```python
def compute_SMA_10(df):
    """
    Вычисляет простое скользящее среднее (SMA) с периодом 10 на основе столбца 'close'.
    """
    df = df.copy()
    df['SMA_10'] = df['close'].rolling(window=10).mean()
    return df

def compute_EMA_10(df):
    """
    Вычисляет экспоненциальное скользящее среднее (EMA) с периодом 10 на основе столбца 'close'.
    """
    df = df.copy()
    df['EMA_10'] = df['close'].ewm(span=10, adjust=False).mean()
    return df

def compute_SMA_50(df):
    """
    Вычисляет простое скользящее среднее (SMA) с периодом 50 на основе столбца 'close'.
    """
    df = df.copy()
    df['SMA_50'] = df['close'].rolling(window=50).mean()
    return df

def compute_EMA_50(df):
    """
    Вычисляет экспоненциальное скользящее среднее (EMA) с периодом 50 на основе столбца 'close'.
    """
    df = df.copy()
    df['EMA_50'] = df['close'].ewm(span=50, adjust=False).mean()
    return df

def apply_moving_averages(df):
    """
    Последовательно применяет все 4 функции вычисления скользящих средних к DataFrame.
    """
    df = compute_SMA_10(df)
    df = compute_EMA_10(df)
    df = compute_SMA_50(df)
    df = compute_EMA_50(df)
    return df


from IPython.display import display

for ticker in tickers:
    var_name = f"df_{ticker}_hourly"
    if var_name in globals():
        df = globals()[var_name]
        df_ma = apply_moving_averages(df)
        globals()[var_name] = df_ma
        print(f"Первые 5 строк часовых данных для {ticker} после добавления SMA_10, EMA_10, SMA_50 и EMA_50:")
        display(df_ma.head(5))
    else:
        print(f"Часовые данные для {ticker} не найдены!")
```

```python
def compute_MACD(df, short_span=12, long_span=26):
    """
    Вычисляет MACD как разницу между EMA с периодом 12 и EMA с периодом 26 на основе 'close'.
    """
    df = df.copy()
    ema_short = df['close'].ewm(span=short_span, adjust=False).mean()
    ema_long = df['close'].ewm(span=long_span, adjust=False).mean()
    df['MACD'] = ema_short - ema_long
    return df

def compute_MACD_signal(df, signal_span=9):
    """
    Вычисляет сигнальную линию MACD (EMA от MACD с периодом 9) и сохраняет в столбце 'MACD_signal'.
    """
    df = df.copy()
    # Предполагается, что столбец 'MACD' уже вычислен
    df['MACD_signal'] = df['MACD'].ewm(span=signal_span, adjust=False).mean()
    return df

def compute_MACD_diff(df):
    """
    Вычисляет разницу между MACD и MACD_signal и сохраняет результат в столбце 'MACD_diff'.
    """
    df = df.copy()
    df['MACD_diff'] = df['MACD'] - df['MACD_signal']
    return df


def compute_ATR_14(df, period=14):
    """
    Вычисляет Average True Range (ATR) за период (по умолчанию 14)
    На основе столбцов 'high', 'low', 'close'.
    True Range (TR) рассчитывается как максимум из:
      - high - low
      - abs(high - prev_close)
      - abs(low - prev_close)
    """
    df = df.copy()
    df['prev_close'] = df['close'].shift(1)
    df['H-L'] = df['high'] - df['low']
    df['H-PC'] = abs(df['high'] - df['prev_close'])
    df['L-PC'] = abs(df['low'] - df['prev_close'])
    df['TR'] = df[['H-L', 'H-PC', 'L-PC']].max(axis=1)
    df['ATR_14'] = df['TR'].rolling(window=period).mean()
    # Удаляем временные столбцы
    df.drop(['prev_close', 'H-L', 'H-PC', 'L-PC', 'TR'], axis=1, inplace=True)
    return df


def compute_OBV(df):
    """
    Вычисляет On-Balance Volume (OBV) на основе 'close' и 'volume'.
    """
    df = df.copy()
    obv = [0]  # начинаем с 0
    for i in range(1, len(df)):
        if df['close'].iloc[i] > df['close'].iloc[i-1]:
            obv.append(obv[-1] + df['volume'].iloc[i])
        elif df['close'].iloc[i] < df['close'].iloc[i-1]:
            obv.append(obv[-1] - df['volume'].iloc[i])
        else:
            obv.append(obv[-1])
    df['OBV'] = obv
    return df

def compute_VWAP(df, window=20):
    """
    Вычисляет Volume Weighted Average Price (VWAP) по скользящему окну.
    Типичная цена = (high + low + close) / 3.
    VWAP = сумма(typical_price * volume) / сумма(volume)
    """
    df = df.copy()
    df['typical_price'] = (df['high'] + df['low'] + df['close']) / 3.0
    df['VWAP'] = (df['typical_price'] * df['volume']).rolling(window=window).sum() / df['volume'].rolling(window=window).sum()
    df.drop('typical_price', axis=1, inplace=True)
    return df


def compute_rolling_std_20(df, window=20):
    """
    Вычисляет скользящее стандартное отклонение от 'close' за 20 периодов.
    """
    df = df.copy()
    df['rolling_std_20'] = df['close'].rolling(window=window).std()
    return df


def apply_hourly_indicators(df):
    """
    Применяет последовательно все функции:
      - MACD, MACD_signal, MACD_diff
      - ATR_14
      - OBV
      - VWAP
      - rolling_std_20
    """
    df = df.copy()
    df = compute_MACD(df)
    df = compute_MACD_signal(df)
    df = compute_MACD_diff(df)
    df = compute_ATR_14(df)
    df = compute_OBV(df)
    df = compute_VWAP(df)
    df = compute_rolling_std_20(df)
    return df


from IPython.display import display

for ticker in tickers:
    var_name = f"df_{ticker}_hourly"
    if var_name in globals():
        df = globals()[var_name]
        df_updated = apply_hourly_indicators(df)
        globals()[var_name] = df_updated
        print(f"Первые 5 строк часовых данных для {ticker} после добавления индикаторов:")
        display(df_updated.head(5))
    else:
        print(f"Часовые данные для {ticker} не найдены!")
```

```python
import numpy as np
import pandas as pd

def compute_Williams_R_hourly(df, period=14):
    """
    Вычисляет Williams %R для часовых данных.
    Формула: -100 * (Highest_High - close) / (Highest_High - Lowest_Low)
    """
    df = df.copy()
    highest_high = df['high'].rolling(window=period).max()
    lowest_low   = df['low'].rolling(window=period).min()
    df['Williams_%R_hourly'] = -100 * (highest_high - df['close']) / (highest_high - lowest_low)
    return df

def compute_Price_deviation_SMA50_hourly(df, window=50):
    """
    Вычисляет 50-часовую SMA для close и процентное отклонение цены закрытия от неё.
    """
    df = df.copy()
    sma50 = df['close'].rolling(window=window).mean()
    df['Price_deviation_SMA50_hourly'] = (df['close'] - sma50) / sma50 * 100
    return df

def compute_HourlyReturn(df):
    """
    Вычисляет процентное изменение цены закрытия (close) между текущим часом и предыдущим.
    """
    df = df.copy()
    df['HourlyReturn'] = df['close'].pct_change() * 100
    return df

def compute_Momentum_5_hourly(df, period=5):
    """
    Вычисляет моментум за 5 часов как разницу между текущим close и close 5 периодов назад.
    """
    df = df.copy()
    df['Momentum_5_hourly'] = df['close'] - df['close'].shift(period)
    return df

def compute_RollingVol_10_hourly(df, window=10):
    """
    Вычисляет скользящее стандартное отклонение процентных изменений close за 10 часов, умноженное на 100.
    """
    df = df.copy()
    df['RollingVol_10_hourly'] = df['close'].pct_change().rolling(window=window).std() * 100
    return df

def compute_RelativeVolume_hourly(df, window=20):
    """
    Вычисляет относительный объём
    """
    df = df.copy()
    df['RelativeVolume_hourly'] = df['volume'] / df['volume'].rolling(window=window).mean()
    return df

def compute_RSI_hourly(df, period=14):
    """
    Вычисляет RSI (Relative Strength Index) для часовых данных по столбцу 'close' с периодом 14.
    """
    df = df.copy()
    delta = df['close'].diff()
    gain = delta.clip(lower=0)
    loss = -delta.clip(upper=0)
    avg_gain = gain.rolling(window=period).mean()
    avg_loss = loss.rolling(window=period).mean()
    rs = avg_gain / avg_loss
    df['RSI'] = 100 - (100 / (1 + rs))
    return df

def apply_hourly_new_indicators(df):
    """
    Применяет все новые индикаторы к часовым данным:
     - Williams_%R_hourly
     - Price_deviation_SMA50_hourly
     - HourlyReturn
     - Momentum_5_hourly
     - RollingVol_10_hourly
     - RelativeVolume_hourly
     - RSI (hourly)
    """
    df = df.copy()
    df = compute_Williams_R_hourly(df, period=14)
    df = compute_Price_deviation_SMA50_hourly(df, window=50)
    df = compute_HourlyReturn(df)
    df = compute_Momentum_5_hourly(df, period=5)
    df = compute_RollingVol_10_hourly(df, window=10)
    df = compute_RelativeVolume_hourly(df, window=20)
    df = compute_RSI_hourly(df, period=14)
    return df


from IPython.display import display

for ticker in tickers:
    var_name = f"df_{ticker}_hourly"
    if var_name in globals():
        df = globals()[var_name]
        df_updated = apply_hourly_new_indicators(df)
        globals()[var_name] = df_updated  
        print(f"\nПервые 5 строк часовых данных для {ticker} после добавления новых индикаторов:")
        display(df_updated.head(5))
    else:
        print(f"Часовые данные для {ticker} не найдены!")
```
